# Thực hành phân tích khám phá dữ liệu với sweetviz

-**SweetViz** là thư viện Python mã nguồn mở giúp tạo các trực quan hóa mật độ cao và báo cáo EDA (Exploratory Data Analysis) chỉ với vài dòng mã.

-những tính năng nổi bật mà SweetViz cung cấp:

* Tự động phân loại kiểu dữ liệu:
SweetViz tự phát hiện biến số là numerical, categorical, hay text mà bạn không cần cấu hình thủ công.

* Thống kê mô tả đầy đủ:
Các chỉ số như min, max, trung bình, phân vị, độ lệch chuẩn, kurtosis, skewness, tổng, etc

* Phân tích mục tiêu (Target Analysis):
Nếu bạn khai báo biến mục tiêu (target), SweetViz sẽ phân tích mối quan hệ giữa các biến khác và biến mục tiêu.

* So sánh hai tập dữ liệu:
Có thể so sánh hai DataFrame (ví dụ: tập train vs test) để xem sự khác biệt, phân phối, giá trị thiếu, v.v.

* So sánh nội tại (intra-comparison):
So sánh các phân nhóm trong cùng tập dữ liệu (ví dụ: nhóm nam vs nữ).

* Hiển thị báo cáo HTML tương tác:
Kết quả được xuất ra dưới dạng HTML có tương tác (nút bấm, điều hướng) để bạn dễ xem và phân tích.

* Tùy chỉnh giao diện báo cáo:
Một số tham số như layout (“widescreen” hay “vertical”), scale, và các tùy chọn hiển thị trong notebook.
