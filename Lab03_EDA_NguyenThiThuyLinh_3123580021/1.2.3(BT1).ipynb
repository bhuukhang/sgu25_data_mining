# Histogram – Phân bố dữ liệu

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

diabetes = pd.read_csv("C:/Users/Admin/Downloads/diabetes.csv")
diabetes['Diagnosis'] = diabetes['Outcome'].map({0: 'Healthy', 1: 'Diabetic'})

# Histogram
plt.figure(figsize=(14,6))

plt.subplot(1,2,1)
sns.histplot(data=diabetes, x='Glucose', bins=25, kde=True, color='salmon')
plt.title('Phân bố mức Glucose', fontsize=14)
plt.xlabel('Glucose')
plt.ylabel('Tần suất')

plt.subplot(1,2,2)
sns.histplot(data=diabetes, x='BMI', bins=25, kde=True, color='skyblue')
plt.title('Phân bố chỉ số BMI', fontsize=14)
plt.xlabel('BMI')
plt.ylabel('Tần suất')

plt.tight_layout()
plt.show()
```

# Scatter Plot – Mối quan hệ giữa Glucose và BMI

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=diabetes, x='BMI', y='Glucose', hue='Diagnosis', palette='coolwarm', alpha=0.7)
plt.title('Mối quan hệ giữa BMI và Glucose', fontsize=14)
plt.xlabel('BMI')
plt.ylabel('Glucose')
plt.show()
```

# Boxplot – So sánh giữa hai nhóm bệnh

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(14,6))

# Subplot 1: Boxplot for Glucose
plt.subplot(1, 2, 1)
sns.boxplot(data=diabetes, x='Diagnosis', y='Glucose', hue='Diagnosis', palette='pastel', legend=False)
plt.title('So sánh Glucose giữa hai nhóm bệnh')

# Subplot 2: Boxplot for BMI
plt.subplot(1, 2, 2)
sns.boxplot(data=diabetes, x='Diagnosis', y='BMI', hue='Diagnosis', palette='muted', legend=False)
plt.title('So sánh BMI giữa hai nhóm bệnh')

plt.tight_layout()
plt.show()
```

# Bar chart – So sánh tuổi trung bình giữa hai nhóm

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6,5))
sns.barplot(data=diabetes, x='Diagnosis', y='Age', hue='Diagnosis', palette='mako', legend=False)
plt.title('Tuổi trung bình theo nhóm bệnh', fontsize=14)
plt.xlabel('Nhóm bệnh')
plt.ylabel('Tuổi trung bình')
plt.show()
```

# Plotly – Biểu đồ tương tác

```python
import plotly.express as px
import plotly.io as pio
pio.renderers.default = 'notebook_connected'

# Biểu đồ tán xạ 3D
fig = px.scatter_3d(
    diabetes,
    x='Glucose',
    y='BMI',
    z='Age',
    color='Diagnosis',
    title='Mối quan hệ 3D giữa Glucose - BMI - Age',
    color_discrete_map={'Healthy': 'blue', 'Diabetic': 'red'}
)
fig.show()

# Biểu đồ cột trung bình
avg_df = diabetes.groupby('Diagnosis')[['Glucose', 'BMI', 'Age']].mean().reset_index()
fig2 = px.bar(
    avg_df,
    x='Diagnosis',
    y=['Glucose', 'BMI', 'Age'],
    barmode='group',
    title='So sánh trung bình Glucose, BMI và Age giữa hai nhóm'
)
fig2.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Tạo biến tỷ lệ Glucose/BMI
diabetes_data['Glucose_BMI_Ratio'] = diabetes_data['Glucose'] / diabetes_data['BMI']

# Sắp xếp theo Glucose giảm dần
diabetes_sorted = diabetes_data.sort_values('Glucose', ascending=False)

# Lấy 10 người đầu tiên
data = diabetes_sorted[0:10]

# case 3: view multiple perspectives at once
fig, ax = plt.subplots(1, 2, figsize=(20, 9))
sns.set(font_scale=1.5)

# Biểu đồ 1 – Mức Glucose
ax1 = sns.barplot(data=data, x='Age', y='Glucose', ax=ax[0], color='salmon', edgecolor='black')
ax1.set_xlabel('Tuổi bệnh nhân (Age)', fontsize=14)
ax1.set_ylabel('Mức đường huyết (Glucose)', fontsize=14)
ax1.set_title('Top 10 bệnh nhân có mức Glucose cao nhất', fontsize=16)

# Biểu đồ 2 – Tỉ lệ Glucose/BMI
ax2 = sns.barplot(data=data, x='Age', y='Glucose_BMI_Ratio', ax=ax[1], color='skyblue', edgecolor='black')
ax2.set_xlabel('Tuổi bệnh nhân (Age)', fontsize=14)
ax2.set_ylabel('Tỉ lệ Glucose/BMI', fontsize=14)
ax2.set_title('Top 10 bệnh nhân có tỉ lệ Glucose/BMI cao nhất', fontsize=16)

plt.tight_layout()
plt.show()
```
