## 1.5. Tổng kết lại

```python
transactions_g5 = df[
    [
        # Delivery
        'late_item',
        'delivery_speed',

        # Product
        'category_item',
        'weight_level',

        # Cost
        'price_level',
        'freight_level',

        # Region
        'state_item',

        # Output
        'review_item'
    ]
].values.tolist()
```

```python
transactions_g5[:5]
```

```python
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
te_array = te.fit(transactions_g5).transform(transactions_g5)

df_trans_g5 = pd.DataFrame(te_array, columns=te.columns_)
```

```python
from mlxtend.frequent_patterns import apriori

frequent_itemsets_g5 = apriori(
    df_trans_g5,
    min_support=0.03,
    use_colnames=True
)
```

```python
from mlxtend.frequent_patterns import association_rules

rules_g5 = association_rules(
    frequent_itemsets_g5,
    metric='confidence',
    min_threshold=0.6
)
```

```python
def count_groups(items):
    groups = {
        'delivery': ['Late_delivery', 'On_time_delivery', 'Speed_'],
        'product': ['Category_', 'Light', 'Medium', 'Heavy'],
        'cost': ['Cheap', 'Medium', 'Expensive', 'Low', 'High'],
        'region': ['State_']
    }

    used = set()
    for item in items:
        for g, keys in groups.items():
            if any(k in item for k in keys):
                used.add(g)
    return len(used)

rules_g5_filtered = rules_g5[
    rules_g5['consequents'].apply(lambda x: any('review' in i.lower() for i in x))
]

rules_g5_filtered['num_groups'] = rules_g5_filtered['antecedents'].apply(count_groups)

rules_g5_final = rules_g5_filtered[
    (rules_g5_filtered['num_groups'] >= 3) &
    (rules_g5_filtered['lift'] > 1.2) &
    (rules_g5_filtered['support'] > 0.01)
]
```

```python
rules_g5_final = rules_g5_final.sort_values(
    by=['lift', 'confidence'],
    ascending=False
)

rules_g5_final.head(10)
```
